```python
import pandas as pd 
import mysql.connector as sqlcon
from sqlalchemy import create_engine  #pip install sqlalchemy
import pymysql  #pip install pymysql
from tabulate import tabulate



#to search a name and delete its record from table
def deletecon():
    try:
        con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="ip2021")
        dname=input("Enter name to delete its Record ")
        qry="delete from contact where name='"+dname+"'"
        cur=con.cursor()
        cur.execute(qry)
        con.commit()
    except:
        print("Delcon: Exception Handling Error")

        

#to add name and contact in using lists
def addcon():
    con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="ip2021")
    cur=con.cursor() 
    try:
        ch=' '
        while ch!='n':
            vname=input("Enter Name of Person")
            vmobile=input("Enter Mobile Number of "+vname)
            qry=("insert into contact(name,mobile) values(%s,%s)")
            data=(vname,vmobile)
            cur.execute(qry,data)
            con.commit()
            ch=input("Add More Records (y/n)?")
    except:
        con.rollback()
        print("addCon Error: while adding data to database/table..")
    
#to add lists in dictionary and display all contacts through dataframe
def dispcon():
    con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="ip2021")
    cur=con.cursor() 
    qry="select * from contact"
    cur.execute(qry)
    data=cur.fetchall()
    headers=["NAME","MOBILE NO."]
    print(tabulate(data,headers=headers))
    # for record in data:
    #     print(record[0],record[1])

#to seach and name and its mobile number 
def searchcon():
     
    con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="ip2021")
    cur=con.cursor()
    sname=input("Enter a name to find Mobile Number ")
    qry="select * from contact where name='%s';"%(sname)
    cur.execute(qry)
    data=cur.fetchall()
    for record in data:
        print(record[0],record[1])
    
#to search a name and modify mobile number
def modifycon():
    try:
        con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="ip2021")
        cur=con.cursor()
        sname=input("Enter a name to modify Mobile Number")
        mobile=input("Enter new Mobile Number ")
        #qry="update contact set mobile='%s' "+(mobile)+" where name='%s';"%(sname)
        qry="update contact set mobile = '%s' where name = '%s';"%(mobile,sname)
        cur.execute(qry)
        con.commit()
        input("Updated Data, Press a key to continue...")
    except:
        print("modify: Record Not Found")


#loop to display a menu of options & execute required above function
ch=' '
while ch!='0':
    print('''
    CONTACTS KEEPER MENU
    0. EXIT
    1. ADD NEW CONTACT
    2. DISPLAY ALL CONTACTS
    3. SEARCH A NAME FROM CONTACT
    4. MODIFY A MOBILE NUMBER
    5. DELETE A RECORD

    ''')
    ch=input("Enter Your Choice 0-5 ")
    if ch=='1':
        addcon()  #calling addcon() function
    elif ch=='2':
        dispcon()   #calling dispcon() function
    elif ch=='3':
        searchcon()  #calling searchcon() function
    elif ch=='4':
        modifycon() #calling modifycon() function
    elif ch=='5':
        deletecon() #calling delcon() function
```
