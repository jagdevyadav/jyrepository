```python
#September 02, 2020
#Framing Flexible Queries with User Data
import pandas as pd
import mysql.connector as sqlcon  #to establish connectivity between Python and MySQL

con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="IP2020")

if con.is_connected():
    mks=float(input("Enter marks: "))
    qry="select * from student where marks>%s;"%(mks,) #%s means there will be a variable 
    df=pd.read_sql(qry,con)
    print("Student details with marks > ",mks)
    print(df)
    con.close()
else:
    print("Sorry! Connection Problem")
```

```python
import pandas as pd
import mysql.connector as sqlcon  #to establish connectivity between Python and MySQL

con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="IP2020")

if con.is_connected():
    lmks=float(input("Enter Lower limit of marks "))
    hmks=float(input("Enter higher limit of marks "))
    qry="select * from student where marks between %s and %s;"%(lmks,hmks) 
    df=pd.read_sql(qry,con)
    print("Student details with marks in range ",lmks," and ",hmks)
    print(df)
    con.close()
else:
    print("Sorry! Connection Problem")
```

```python
import pandas as pd
import mysql.connector as sqlcon  #to establish connectivity between Python and MySQL

con=sqlcon.connect(host="localhost",user="root",passwd="tiger",database="IP2020")

if con.is_connected():
    letter=input("Enter a letter to search name starting with it ")
    qry="select * from student where name like '%s%%';"%(letter) 
    df=pd.read_sql(qry,con)
    print("Student details whose name beginning with letter ",letter)
    print(df)
    con.close()
else:
    print("Sorry! Connection Problem")
```
